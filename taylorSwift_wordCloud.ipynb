```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download("stopwords")
nltk.download('punkt')
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# Load the dataset
with open('/content/taylor.txt', 'r') as file:
    text = file.readlines()

# Assuming each line is a single data point, you can create a DataFrame like this:
df = pd.DataFrame(text)
```

```python
# Preprocessing - changes document to lower case and removes stopwords
def preprocess(document):

    # change sentence to lower case
    document = document.lower()

    # tokenize into words
    words = word_tokenize(document)

    # remove stop words
    words = [word for word in words if word not in stopwords.words("english")]

    # join words to make sentence
    document = " ".join(words)

    return document
```

```python
# Each sentence as a list item (preprocessed)
text = [preprocess(text) for text in text]
print(text)
```

```python
# making the text doc into one huge blob
corpus = " ".join(text)
#splitting it into seperate words
corpus.split()
```

```python
#tokenzing
words=word_tokenize(corpus)
print(words)
```

```python
#removing stopwords
words = [word for word in words if word not in stopwords.words('english')]
print(words)
```

```python
# join it all bak
text = " ".join(words)
print(text)
```

```python
# making a word cloud
stopwords = set(STOPWORDS)
wordcloud = WordCloud(max_font_size=40, max_words=200, background_color="black", stopwords=stopwords).generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
